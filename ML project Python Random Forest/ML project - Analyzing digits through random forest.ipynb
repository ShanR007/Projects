```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df[0:12]
```

```python
df['target'] = digits.target
df.head()
# These targets all map to a certain value, for eg. for the index 1 the value is o 
```

```python
# Training and testing the model for prediction

X = df.drop('target',axis='columns')
y = df.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
# ensemble since we ae using multiple algorithms to predict the outcome, 
# we are building multiple decision trees and to commbine and build a prediction
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
# Training step
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
# Making the confusion matrix
# to find the errors, and where the model performs well in prediction and where it does not
# perform well or where it performs poorly
# It plots 2D array, truth on one axis and the prediction on the other axis
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
# cm is the confusion matrix
```

```python
# using seaborn visualization library to visualise the above matrix
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
